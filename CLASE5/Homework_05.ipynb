```python
#OPCION 2
import random as r

class JuegoPila:
    def __init__(self) :
        self.__lista = []

    #Agrega un elemento a la pila
    def __push(self, item):
        return self.__lista.append(item)

    #Quitar un elemto de la pila
    def __pop(self):
        return self.__lista.pop()

    #Mostrar la pila
    def __mostrarPila(self):
        print(self.__lista) 

    def jugar(self):
        while len(self.__lista) < 20:
            numero = r.randint(1, 20)
            if numero not in self.__lista: self.__push(numero)
        #Otra version para crear la pila
        #self.__lista = r.sample(range(1, 21), 20)
        nro = input("Ingrese la cantidad de elementos a retirar: ")
        while not nro.isdecimal() or int(nro) > 20:
            nro = input("Ingrese la cantidad de elementos a retirar nuevamente (tiene que ser un numero entero 1-20): ")
        nro = int(nro)

        print("Pila original")
        self.__mostrarPila()
        print("Cantidad de elementos a retirar: ", nro)

        calificacion = 10
        suma = 0
        for i in range(nro):
            suma += self.__pop()

        print("Pila obtenida")
        self.__mostrarPila()
        print("Cantidad de elementos quitados suma: ", suma)

        if suma > 50: print('Perdiste!!! Te pasaste del valor de 50')
        else:
            while suma <= 50:
                suma += self.__pop()
                if suma <= 50: calificacion -= 1
            print("Felicitaciones!!! Ganaste, tu puntaje es: ", calificacion)
```

2) Implementar un juego donde constas de 2 jarras, de capacidad 5 y 3 litros respectivamente, y debes colocar 4 litros en la jarra de 5L.
Las opciones posibles son:
* Llenar la jarra de 3 litros
* Llenar la jarra de 5 litros
* Vaciar la jarra de 3 litros
* Vaciar la jarra de 5 litros
* Verter el contenido de la jarra de 3 litros en la de 5 litros
* Verter el contenido de la jarra de 5 litros en la de 3 litros

```python
class Jarras:
    def __init__(self):
        self.jarra_5L = 0
        self.jarra_3L = 0

    def llenar_jarra_5L(self):
        self.jarra_5L = 5

    def llenar_jarra_3L(self):
        self.jarra_3L = 3

    def vaciar_jarra_5L(self):
        self.jarra_5L = 0

    def vaciar_jarra_3L(self):
        self.jarra_3L = 0

    def verter_3L_en_5L(self):
        espacio_disponible = 5 - self.jarra_5L
        cantidad_a_verter = min(self.jarra_3L, espacio_disponible)
        self.jarra_5L += cantidad_a_verter
        self.jarra_3L -= cantidad_a_verter

    def verter_5L_en_3L(self):
        espacio_disponible = 3 - self.jarra_3L
        cantidad_a_verter = min(self.jarra_5L, espacio_disponible)
        self.jarra_3L += cantidad_a_verter
        self.jarra_5L -= cantidad_a_verter

    def estado_actual(self):
        return f"Jarra de 5L: {self.jarra_5L} litros | Jarra de 3L: {self.jarra_3L} litros"


def juego():
    jarras = Jarras()

    while jarras.jarra_5L != 4:
        print(jarras.estado_actual())

        print("Opciones:")
        print("1. Llenar la jarra de 3 litros")
        print("2. Llenar la jarra de 5 litros")
        print("3. Vaciar la jarra de 3 litros")
        print("4. Vaciar la jarra de 5 litros")
        print("5. Verter el contenido de la jarra de 3 litros en la de 5 litros")
        print("6. Verter el contenido de la jarra de 5 litros en la de 3 litros")

        opcion = int(input("Selecciona una opción (1-6): "))

        if opcion == 1:
            jarras.llenar_jarra_3L()
        elif opcion == 2:
            jarras.llenar_jarra_5L()
        elif opcion == 3:
            jarras.vaciar_jarra_3L()
        elif opcion == 4:
            jarras.vaciar_jarra_5L()
        elif opcion == 5:
            jarras.verter_3L_en_5L()
        elif opcion == 6:
            jarras.verter_5L_en_3L()
        else:
            print("Opción no válida. Inténtalo de nuevo.")

    print("¡Ganaste! Has logrado colocar 4 litros en la jarra de 5 litros.")

if __name__ == "__main__":
    juego()
```
